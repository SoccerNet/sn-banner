```python
import os
from PIL import Image
import numpy as np
from copy import deepcopy
from matplotlib import pyplot as plt
from tqdm import tqdm
```

```python
# if not yet done, change directory to 'Datasetv2' folder
if os.getcwd().split('/')[-1] != 'Datasetv3':
    os.chdir('Datasetv3')
    print('Changed directory to Datasetv2')
else:
    print('Current directory: ', os.getcwd())
```

```python
old_classes = ['Outside billboards', 'Billboard', 'Field player', 'Goalkeeper', 'Referee', 'Assistant referee', 'Other human', 'Ball', 'Goal post', 'Goal net', 'Net post', 'Cross-bar', 'Corner flag', 'Assistant flag', 'Microphone', 'Camera', 'Other object', 'Don\'t care']
new_classes = ['Outside billboards', 'Billboard', 'Goal net', 'Occlusion']
conversion = np.array([new_classes.index('Occlusion')] * len(old_classes), dtype=np.uint8)  # default to 'Occlusion'
conversion[old_classes.index('Outside billboards')] = new_classes.index('Outside billboards')
conversion[old_classes.index('Billboard')] = new_classes.index('Billboard')
conversion[old_classes.index('Goal net')] = new_classes.index('Goal net')
old_palette = [[0, 0, 0], [255, 255, 255], [255, 0, 0], [0, 255, 0], [0, 0, 255], [255, 255, 0], [255, 0, 255], [0, 255, 255], [128, 0, 0], [0, 128, 0], [0, 0, 128], [64, 64, 64], [128, 128, 0], [128, 0, 128], [0, 128, 128], [255, 128, 0], [192, 192, 192], [128, 128, 128]]
new_palette = [[0, 0, 0], [255, 255, 255], [0, 255, 0], [255, 0, 0]]
```

```python
original_labels = os.listdir('Labels_source/')
print('Number of original labels: ', len(original_labels))
```

```python
for label in tqdm(original_labels):
	img = Image.open('Labels_source/' + label)
	img = np.array(img, dtype=np.uint8)
	img = conversion[img]
	img = Image.fromarray(img).convert('P')
	img.putpalette(np.array(new_palette, dtype=np.uint8))
	img.save('Labels/' + label)
```
