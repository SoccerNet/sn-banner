```python
import os
```

```python
# List files in Images/ directory
files = os.listdir("Images/")
print(files)
print(len(files))
```

```python
train = set(["Stadium_4_Match_1", "Stadium_5_Match_1", "Stadium_6_Match_2"])
print(train)
# The train split is composed of match 1 of stadium 4, 1 of 5 and 2 of 6
# The validation split is composed of 2 of 4, 1 of 3 and 1 of 2
val = set(["Stadium_4_Match_2", "Stadium_3_Match_1", "Stadium_2_Match_1"])
# The test split is composed of 2 of 5, 1 of 6 and 1 of 8
test = set(["Stadium_5_Match_2", "Stadium_6_Match_1", "Stadium_8_Match_1"])
# The challenge split is composed of 1 of 7 and 2 of 1
challenge = set(["Stadium_7_Match_1", "Stadium_1_Match_2"])
# The private split is composed of 1, 3, 4 of stadium 1
private = set(["Stadium_1_Match_1", "Stadium_1_Match_3", "Stadium_1_Match_4"])
```

```python
split_dir = "splits/"
```

```python
# Create and open the files for each split
train_file = open(split_dir + "train.txt", "w")
val_file = open(split_dir + "val.txt", "w")
test_file = open(split_dir + "test.txt", "w")
challenge_file = open(split_dir + "challenge.txt", "w")
private_file = open(split_dir + "private.txt", "w")
```

```python
# Each image begins with "Stadium_#_Match_#" where # is a number
# For each file in files, get the beginning of the filename

for file in files:
    # Get the 17 first characters of the filename
    file_split = file[:17]
    # Rename the file without the .png extension
    file_renamed = file[:-4]
    if file_split in train:
        train_file.write(file_renamed + "\n")
    elif file_split in val:
        val_file.write(file_renamed + "\n")
    elif file_split in test:
        test_file.write(file_renamed + "\n")
    elif file_split in challenge:
        challenge_file.write(file_renamed + "\n")
    elif file_split in private:
        private_file.write(file_renamed + "\n")
    else:
        print("Error: " + file_split + " not in any split")
```

```python
# Close the files
train_file.close()
val_file.close()
test_file.close()
challenge_file.close()
private_file.close()
```
