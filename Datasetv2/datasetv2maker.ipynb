```python
import os
from PIL import Image
import numpy as np
from copy import deepcopy
from matplotlib import pyplot as plt
from tqdm import tqdm
```

```python
# if not yet done, change directory to 'Datasetv2' folder
if os.getcwd().split('/')[-1] != 'Datasetv2':
    os.chdir('Datasetv2')
    print('Changed directory to Datasetv2')
```

```python
# list label images in the dir 'Labels'
labels = os.listdir('Labels')
print('Found', len(labels), 'label images')
```

```python
images_without_gl = []
for label in tqdm(labels):
    # open label image
    img = Image.open('Labels/' + label)
    # The image is a 2d array with values 0-2, i.e. a class index
    # Convert to numpy array
    img = np.array(img)
    bb = deepcopy(img)
    # Change all 2 values to 0
    bb[bb == 2] = 0
    # Save bb image in 'Billboard/Labels' folder
    Image.fromarray(bb).save('Billboard/Labels/' + label)

    # Check if there is at least one 2 value in the image
    if 2 in img:
        gl = deepcopy(img)
        # Change all 1 values to 0
        gl[gl == 1] = 0
        # Change all 2 values to 1
        gl[gl == 2] = 1
        # Save gl image in 'Goal/Labels' folder
        Image.fromarray(gl).save('Goal/Labels/' + label)
    else:
        images_without_gl.append(label)
```
